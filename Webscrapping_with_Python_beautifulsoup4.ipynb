### Getting the HTML directly from the URL

```
import requests

url = 'https://www.ohoshop.in/features/'
r = requests.get(url)
r.text
```

```
soup = BeautifulSoup(r.text, 'html.parser')
# print(soup.prettify())
```

```
features = []

for link in soup.find_all(['h3','p']):
  features.append(link.get_text())
```

```
feat = []
indx = []
for i in features[0:len(features):2]:
  # print(i)
  feat.append(i)

```

```
count = 0
for i in features[1:len(features):2]:
  # print(i)
  indx.append(i)
  count += 1

print(count)
```
