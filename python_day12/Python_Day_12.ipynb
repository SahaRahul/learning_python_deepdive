```
import matplotlib.pyplot as plt
import numpy as np
```

```
X = [1,2,3,4]
p1 = [50, 60, 70, 80]
p2 = [30, 60, 20, 90]
p3 = [20, 70, 90, 100]

plt.plot(X, p1, 'g', label="Product 1", linewidth=3)
plt.plot(X, p2, 'c', label="Product 2", linewidth=3)
plt.plot(X, p3, 'y', label="Product 3", linewidth=3)

plt.title("Product performance in four quater")
plt.xlabel("Nos of Quater")
plt.ylabel("Product Sales")
plt.legend()
plt.show()
```

```
p1 = [50, 20, 20, 10]
p2 = [30, 20, 20, 30]
p3 = [20, 60, 60, 60]

plt.bar(X,p1, label="Prod 1", width=0.5)
plt.bar(X,p2, label="Prod 2", width=0.5)
plt.bar(X,p3, label="Prod 3", width=0.5)

plt.title("Product performance in four quater")
plt.xlabel("Nos of Quater")
plt.ylabel("Product Sales")
plt.legend()
plt.figure(figsize=(8,6), dpi=80,edgecolor='k')
plt.show()
```

```
points  = [1, 2, 3, 4, 5]
values = [1, 4, 9, 16, 25]

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(points, values, s = 100)
ax.set_title("Square number plot", fontsize=24)
ax.set_xlabel("Value")
ax.set_ylabel("Square of value")
ax.tick_params(axis='both',labelsize=14)
plt.show()
```

```
points  = range(1, 1001)
values = [x**2 for x in points]

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(points, values, s = 100)
ax.set_title("Square number plot", fontsize=24)
ax.set_xlabel("Value")
ax.set_ylabel("Square of value")
ax.tick_params(axis='both',labelsize=14)
plt.show()
```

```
points  = range(1, 1001)
values = [x**2 for x in points]

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(points, values, c = values, cmap=plt.cm.BrBG_r, s = 10)
ax.set_title("Square number plot", fontsize=24)
ax.set_xlabel("Value")
ax.set_ylabel("Square of value")
ax.tick_params(axis='both',labelsize=14)
plt.savefig("practice_plt.png")
plt.show()
```

```
xlist = np.linspace(-3.0,3.0, 100)
ylist = np.linspace(-3.0,3.0, 100)

X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 - Y**2)

fig, ax = plt.subplots(1,1)
# cp = ax.contour(X,Y,Z)
cp = ax.contourf(X,Y,Z)

fig.colorbar(cp)

plt.show()
```
